```python
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark.sql.functions import count
import matplotlib.pyplot as plt 
```

## Create a spark session

```python
aws_access_key_id = ""
aws_secret_access_key = ""
aws_container_url = "s3a://eclipse-storage"

conf = (
    SparkConf()
    .setAppName("MY_APP") # replace with your desired name
    .set("spark.jars.packages", "io.delta:delta-core_2.12:2.3.0,org.apache.hadoop:hadoop-aws:3.3.2")
    .set("spark.sql.catalog.spark_catalog","org.apache.spark.sql.delta.catalog.DeltaCatalog")
    .set("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
    .set("spark.hadoop.fs.s3a.access.key", aws_access_key_id)
    .set("spark.hadoop.fs.s3a.secret.key", aws_secret_access_key)
    .set("spark.sql.shuffle.partitions", "4") # default is 200 partitions which is too many for local
    .setMaster("local[*]")
)

pyspark_session = SparkSession.builder.config(conf=conf).getOrCreate()
```

## Define data urls

```python
ingestion_data_url = f"{aws_container_url}/bronze/ingestion_output"
average_financial_product_url = f"{aws_container_url}/gold/average_finicial_product"
highest_volume_financial_product_url = f"{aws_container_url}/gold/highest_volume_financial_product"
anomalies_detection_financial_product_url = f"{aws_container_url}/gold/anomalies_detection_financial_product"
```

## Ingestion output

```python
ingestion_df = pyspark_session.read.parquet(ingestion_data_url).toPandas()
```

### Display Number of energy trade per hour

```python
plt.title('Number of energy trade per hours')
ingestion_df.groupby(ingestion_df['DeliveryStart'].rename('Hours').dt.hour).size().plot(xlabel='hours', ylabel='Trade count')
```

```python
# Average financial Product ouput
```

```python
average_financial_product_df = pyspark_session.read.parquet(average_financial_product_url).toPandas()
```

```python
# Highest volume financial product output
```

```python
avergae_price_through_day = average_financial_product_df.sort_values(by="DeliveryStart")
avergae_price_through_day.set_index('DeliveryStart', inplace=True)
plt.title('Average price through day')
avergae_price_through_day["AverageTradePrice"].plot()
```

```python
avergae_price_through_day = average_financial_product_df.sort_values(by="DeliveryStart")
avergae_price_through_day.set_index('DeliveryStart', inplace=True)
plt.title('Average Volume through day')
avergae_price_through_day["AverageVolume"].plot()
```
